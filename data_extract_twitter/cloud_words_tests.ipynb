```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
import re
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
import nltk 
from nltk.corpus import stopwords

nltk.download('stopwords')
from nltk.tokenize import word_tokenize
```

```python
stopwords = " | ".join(stopwords.words('portuguese'))
```

```python
df = pd.read_csv("./data_tweets/persist_tweets_2021-03-07-19-13-24-863807_2021-03-07-19-27-29-932784.csv")
```

```python
df.head()
```

```python
text = " ".join(review for review in df.text)

text = re.sub(r"(?i){}".format(stopwords),"",text)

text = re.sub(r"(?i)[^A-Za-z1-9]t[^A-Za-z1-9]|\/co\/|bbb21|oficial|boninho|RedeGlobo|globo|gostaria|solicitar|support|reality|show|@bbb|Big Brother Brasil|Twitter|Brasil|https|TwitterSupport|verificao|verified","",text)
text = text.replace("/co/","")

wordcloud = WordCloud(max_font_size=200, max_words=200, background_color="black",width=3000, height=1300).generate(text)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```
